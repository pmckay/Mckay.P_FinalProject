```python
### import block 
import pandas as pd
import numpy as np
import xarray as xr
from datetime import datetime
from datetime import timedelta
import netCDF4
### imports for graphing
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import cmocean as cm
```

```python
### initiating filepath to north data
filepath_north = '~/project-stuff/data/north.p16.csv'

### naming the filepath north_df, and removing headers as well as first row and last row becasue it was not atual data
north_df = (pd.read_csv(filepath_north, skiprows = 34,  sep = ',')).iloc[1:-1]

#display(north_df)
```

```python
#making variables 

#equator data
# declare these first so that I can round them
north_temp = north_df['CTDTMP'].astype(float)
north_lat = north_df['LATITUDE'].astype(float)

#rounding to nearest .5 values for both lat and temp
# not rounding ph because there are already similar values. 
north_df['round_temp']= np.round(north_temp*2)/2
north_df['round_lat'] = np.round(north_lat*2)/2

# make variables for the new columns created right above and index by correct temp values
north_roundtemp = north_df['round_temp'].loc[north_temp <= 10].astype(float)
north_roundlat = north_df['round_lat'].loc[north_temp <= 10].astype(float)


# making variables that are indexed by the non errored temp and also turn into floats
north_depth = north_df['DEPTH'].loc[north_temp <= 10].astype(float)
north_ph = north_df['PH_TOT'].loc[north_temp <= 10].astype(float)
north_lat = north_df['LATITUDE'].loc[north_temp <= 10].astype(float)
north_lon = north_df['LONGITUDE'].loc[north_temp <= 10].astype(float) #not activly using this but good to have
north_temp = north_df['CTDTMP'].loc[north_temp <= 10].astype(float)

```

```python
# plotting a graph of temp, depth, and lat
# general figure 
fig, ax = plt.subplots(figsize =(15,6))
sp = ax.scatter(north_roundlat, north_depth, c = north_roundtemp, marker = '.', label = 'Temperature', vmin = 0, vmax =4)

# aestetics of plot
ax.invert_yaxis()
ax.set_ylabel('Depth (m)', fontsize = 14)
ax.set_xlabel('Latitude', fontsize = 14)
plt.title('Temperature Variation by Latitude and Depth', fontsize = 14)              
c1 = plt.colorbar(sp, ax=ax )
c1.set_label('Temperature (C)', fontsize = 14)
ax.grid(axis = 'x')

# commenting out because ive already saved the fig
# plt.savefig( 'TEMP 215 graph north')
```

```python
# subset the already subsetted data by ph this time to delete errored PH values
# use loc wiht boolean condition to get only the values that arent -999 error

north_roundlat = north_roundlat.loc[north_ph != -999]
north_depth = north_depth.loc[north_ph != -999]
north_ph = north_ph.loc[north_ph != -999]
```

```python
# plotting a graph of pH, depth, and lat.
# general figure 
fig, ax = plt.subplots(figsize =(15,6))
sp2 = ax.scatter(north_roundlat, north_depth, c= north_ph, cmap = 'magma', marker = '.', label = 'pH', vmin=6.5, vmax = 8.5)

# figure aestetics
ax.invert_yaxis()
ax.set_ylabel('Depth (m)',fontsize = 14)
ax.set_xlabel('Latitude', fontsize = 14)
c2 = plt.colorbar(sp2, ax=ax )
c2.set_label('pH', fontsize = 14)
plt.title('pH Variation by Latitude and Depth ', fontsize = 14)  
ax.grid(axis = 'x')

# commenting out because ive already saved the fig
# plt.savefig('pH 215 graph north')
```
